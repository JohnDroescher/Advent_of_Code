```python
import numpy as np
import pandas as pd

data = pd.DataFrame(np.genfromtxt('input.txt', delimiter=1))
data = data.T
data['gamma'] = (data.sum(axis=1) > 500).astype(int)
data['eps'] = (data.sum(axis=1) < 500).astype(int)
```

```python
def binary_calc(binary):
    gamma = 0
    gamma_mult = 1
    for val in reversed(binary):
        gamma += gamma_mult * val
        gamma_mult *= 2
    return gamma

binary_calc(data.gamma) * binary_calc(data.eps)
```
