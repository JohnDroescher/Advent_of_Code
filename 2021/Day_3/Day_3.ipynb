```python
import numpy as np
import pandas as pd


def binary_calc(binary):
    gamma = 0
    gamma_mult = 1
    for val in reversed(binary):
        gamma += gamma_mult * val
        gamma_mult *= 2
    return gamma


def other_binary(binary):
    binary = "".join([str(v) for v in binary])
    return int(binary, 2)


def o2_co2_generator(input_txt, o2=True):
    data = input_txt.copy()
    if o2:
        for i in range(data.shape[1]):
            # Check for last row
            if data.shape[0] > 1:
                # Check if 1 is most popular value in column
                key_val = data.shape[0] / 2 <= data[i].sum()
                # Keep most popular value
                data = data[data[i] == int(key_val)].drop(i, axis=1)
                # Define index_row as first row idx
                index_row = data.index[0]

            else:
                # If last row remaining, skip to returning row idx
                index_row = data.index[0]

    else:
        for i in range(data.shape[1]):
            # Check for last row
            if data.shape[0] > 1:
                # Check if 1 is popular value in column
                key_val = data.shape[0] / 2 <= data[i].sum()
                # Drop most popular value in column
                data = data[data[i] != int(key_val)].drop(i, axis=1)
                # Define index_row output as first row idx
                index_row = data.index[0]

            else:
                # If last row remaining, skip to returning row idx
                index_row = data.index[0]

    # Run the output through the binary -> int convertor function
    return binary_calc(input_txt.iloc[index_row])
```

```python
# Load the data sets
data = pd.DataFrame(np.genfromtxt("input.txt", delimiter=1))
data2 = pd.DataFrame(np.genfromtxt("test.txt", delimiter=1))

# Step 1
data_step_1 = data.T
data_step_1["gamma"] = (data_step_1.sum(axis=1) > 500).astype(int)
data_step_1["eps"] = (data_step_1.sum(axis=1) < 500).astype(int)

print("Step 1 Answer:")
print(binary_calc(data_step_1.gamma) * binary_calc(data_step_1.eps))
print("Step 1 Answer w/alternate binary calculator")
print(other_binary(data_step_1.gamma) * other_binary(data_step_1.eps))

# Step 2
print("\n" + "Step 2 Answer:")
print(
    int(o2_co2_generator(data, o2=False) * o2_co2_generator(data))
)
```

```python
# From Kyle Palko, to verify my function is working properly

import numpy as np


def frequency_digits_place(a):
    """Find the most frequent number (1 or 0) in a binary digit's place"""
    frequency = np.zeros((len(a[0])))

    for num in a:
        for idx, char in enumerate(num):
            frequency[idx] += int(char)

    return frequency


def frequency_in_binary(data):
    freq = frequency_digits_place(data)
    return [1 if val >= len(data) / 2 else 0 for val in freq]


def invert_binary(a):
    return [1 if val == 0 else 0 for val in a]


def binary_to_int(a):
    """Converts a list or array of binary integers into a string and then an integer"""
    a = "".join([str(v) for v in a])
    return int(a, 2)


def day_3a(data):

    gamma = frequency_in_binary(data)
    epsilon = invert_binary(gamma)
    gamma = binary_to_int(gamma)
    epsilon = binary_to_int(epsilon)

    print(gamma * epsilon)


def day_3b(data):

    # find O2 by iterating through the full array of data
    oxygen_list = data[:]
    i = 0
    while len(oxygen_list) > 1 and i < len(data[0]):

        frequency = frequency_in_binary(oxygen_list)
        tracker = []
        for idx, val in enumerate(oxygen_list):
            if int(val[i]) != frequency[i]:
                tracker.append(idx)
        oxygen_list = np.delete(oxygen_list, tracker)
        i += 1

    # find CO2
    co2_list = data[:]
    i = 0
    while len(co2_list) > 1 and i < len(data[0]):

        frequency = frequency_in_binary(co2_list)
        tracker = []
        for idx, val in enumerate(co2_list):
            if int(val[i]) == frequency[i]:
                tracker.append(idx)
        co2_list = np.delete(co2_list, tracker)
        i += 1

    ans = binary_to_int(oxygen_list) * binary_to_int(co2_list)
    print(ans)


data = np.genfromtxt("input.txt", dtype="str")
data2 = np.genfromtxt("test.txt", dtype="str")
# data = np.genfromtxt('day_03/unit_data.txt', dtype='str', skip_header=1)  # unit test

day_3a(data)
day_3b(data)
```
