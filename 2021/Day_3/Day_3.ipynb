```python
import numpy as np
import pandas as pd


def binary_calc(binary):
    gamma = 0
    gamma_mult = 1
    for val in reversed(binary):
        gamma += gamma_mult * val
        gamma_mult *= 2
    return gamma


def other_binary(binary):
    binary = "".join([str(v) for v in binary])
    return int(binary, 2)


def o2_co2_generator(input_txt, o2=True):
    data = input_txt.copy()
    if o2:
        for i in range(data.shape[1]):
            # Check for last row
            if data.shape[0] > 1:
                # Check if 1 is most popular value in column
                key_val = data.shape[0] / 2 <= data[i].sum()
                # Keep most popular value
                data = data[data[i] == int(key_val)].drop(i, axis=1)
                # Define index_row as first row idx
                index_row = data.index[0]

            else:
                # If last row remaining, skip to returning row idx
                index_row = data.index[0]

    else:
        for i in range(data.shape[1]):
            # Check for last row
            if data.shape[0] > 1:
                # Check if 1 is popular value in column
                key_val = data.shape[0] / 2 <= data[i].sum()
                # Drop most popular value in column
                data = data[data[i] != int(key_val)].drop(i, axis=1)
                # Define index_row output as first row idx
                index_row = data.index[0]

            else:
                # If last row remaining, skip to returning row idx
                index_row = data.index[0]

    # Run the output through the binary -> int convertor function
    return binary_calc(input_txt.iloc[index_row])
```

```python
# Load the data sets
data = pd.DataFrame(np.genfromtxt("input.txt", delimiter=1))
data2 = pd.DataFrame(np.genfromtxt("test.txt", delimiter=1))

# Step 1
data_step_1 = data.T
data_step_1["gamma"] = (data_step_1.sum(axis=1) > 500).astype(int)
data_step_1["eps"] = (data_step_1.sum(axis=1) < 500).astype(int)

print("Step 1 Answer:")
print(binary_calc(data_step_1.gamma) * binary_calc(data_step_1.eps))

print("Step 1 Answer w/alternate binary calculator")
print(other_binary(data_step_1.gamma) * other_binary(data_step_1.eps))

# Step 2
print("\n" + "Step 2 Answer:")
print(int(o2_co2_generator(data, o2=False) * o2_co2_generator(data)))
```
